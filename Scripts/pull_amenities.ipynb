```python
## Libraries and setup
# !pip install diagonal_b6
import diagonal_b6 as b6
import os
import json
import geopandas as gpd
import pandas as pd


# get working directory
working_directory =  os.getcwd()

# connect to the Bedrock engine
w = b6.connect_insecure("cdtweek-2023.diagonal.works:8002")
## get lsoas
# get Bedrock data for Liverpool study area
liverpool = b6.cap_polygon(b6.ll(53.4127,-2.9608), 12000)
```

```python
# get lsoas
lsoas = b6.find(b6.and_(b6.tagged("#boundary", "lsoa"), b6.intersecting(liverpool)))
# ## get pwc
# # set filenames
# #filename = "\pwc_liverpool.geojson"
# filename = "/pwc_liverpool.geojson"

data_directory = working_directory.replace('Scripts', 'Data')

# #  create geodataframe from file
# pwc = gpd.read_file(data_directory + filename)
```

```python
# Download all the amenities of each category in Liverpool
point_y = (-2.9890646186096115)
point_x = (53.40552586245608)

# get area around the city centre of liverpool
temp_area = b6.cap_polygon(b6.ll(point_x, point_y)
                           , 100000 # set this to a large distance that covers the whole city
                           )

# Schools
schools = b6.find(b6.and_(b6.tagged("#amenity","school"), b6.intersecting(temp_area))) # get all the amenities 
schools_geojson = w(b6.to_geojson_collection(schools)) # save results to geojson
schools_gdf = gpd.GeoDataFrame.from_features(schools_geojson) # save geojson to gdf
schools_gdf.to_file(data_directory + '/Data/schools_gdf.geojson') # export to geojson

# Supermarkets
supermarkets = b6.find(b6.and_(b6.tagged("#shop","supermarket"), b6.intersecting(temp_area))) # get all the amenities 
supermarkets_geojson = w(b6.to_geojson_collection(supermarkets)) # save results to geojson
supermarkets_gdf = gpd.GeoDataFrame.from_features(supermarkets_geojson) # save geojson to gdf
supermarkets_gdf.to_file(data_directory + '/Data/supermarkets_gdf.geojson') # export to geojson

# pharmacies
pharmacies = b6.find(b6.and_(b6.tagged("#amenity","pharmacy"), b6.intersecting(temp_area))) # get all the amenities 
pharmacies_geojson = w(b6.to_geojson_collection(pharmacies)) # save results to geojson
pharmacies_gdf = gpd.GeoDataFrame.from_features(pharmacies_geojson) # save geojson to gdf
pharmacies_gdf.to_file(data_directory + '/Data/pharmacies_gdf.geojson') # export to geojson

# atms
atms = b6.find(b6.and_(b6.tagged("#amenity","atm"), b6.intersecting(temp_area))) # get all the amenities 
atms_geojson = w(b6.to_geojson_collection(atms)) # save results to geojson
atms_gdf = gpd.GeoDataFrame.from_features(atms_geojson) # save geojson to gdf
atms_gdf.to_file(data_directory + '/Data/atms_gdf.geojson') # export to geojson

# parks
parks = b6.find(b6.and_(b6.tagged("#leisure","park"), b6.intersecting(temp_area))) # get all the amenities 
parks_geojson = w(b6.to_geojson_collection(parks)) # save results to geojson
parks_gdf = gpd.GeoDataFrame.from_features(parks_geojson) # save geojson to gdf
parks_gdf.to_file(data_directory + '/Data/parks_gdf.geojson') # export to geojson

# cinemas
cinemas = b6.find(b6.and_(b6.tagged("#amenity","cinema"), b6.intersecting(temp_area))) # get all the amenities 
cinemas_geojson = w(b6.to_geojson_collection(cinemas)) # save results to geojson
cinemas_gdf = gpd.GeoDataFrame.from_features(cinemas_geojson) # save geojson to gdf
cinemas_gdf.to_file(data_directory + '/Data/cinemas_gdf.geojson') # export to geojson

# gyms
gyms = b6.find(b6.and_(b6.tagged("#leisure","fitness_centre"), b6.intersecting(temp_area))) # get all the amenities 
gyms_geojson = w(b6.to_geojson_collection(gyms)) # save results to geojson
gyms_gdf = gpd.GeoDataFrame.from_features(gyms_geojson) # save geojson to gdf
gyms_gdf.to_file(data_directory + '/Data/gyms_gdf.geojson') # export to geojson

# playgrounds
playgrounds = b6.find(b6.and_(b6.tagged("#leisure","playground"), b6.intersecting(temp_area))) # get all the amenities
playgrounds_geojson = w(b6.to_geojson_collection(playgrounds)) # save results to geojson
playgrounds_gdf = gpd.GeoDataFrame.from_features(playgrounds_geojson) # save geojson to gdf
playgrounds_gdf.to_file(data_directory + '/Data/playgrounds_gdf.geojson') # export to geojson

# pitches
pitches = b6.find(b6.and_(b6.tagged("#leisure","pitch"), b6.intersecting(temp_area)))# get all the amenities
pitches_geojson = w(b6.to_geojson_collection(pitches)) # save results to geojson
pitches_gdf = gpd.GeoDataFrame.from_features(pitches_geojson) # save geojson to gdf
pitches_gdf.to_file(data_directory + '/Data/pitches_gdf.geojson') # export to geojson
```
